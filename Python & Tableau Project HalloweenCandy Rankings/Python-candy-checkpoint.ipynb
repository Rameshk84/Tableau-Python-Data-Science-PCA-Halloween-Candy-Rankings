```python
## Goal
## Use data to find the three types of Halloween candy that will make you the most popular house on the block

## Approach
## View the data in Python
## Apply PCA to the data to create a scatter plot in Python
## Identify clusters and filter the data down to three candies in Tableau
## My Top Picks
## A popular candy: Reese's Peanut Butter Cup
## A popular chocolate option: Twix
## A popular fruity option: Starburst
```

```python
import pandas as pd 
data = pd.read_csv('C:/Users/Ramesh K/Downloads/Halloween+Candy+Rankings/candy-data.csv')
data.head()
```

```python
subset = data.iloc[:,1:-3]
subset.head()
```

```python
## Apply (PCA) Principle Component analysis
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(subset)
```

```python
pca.explained_variance_ratio_
```

```python
candy_2D = pd.DataFrame(pca.transform(subset))
candy_2D.head()
```

```python
candy_2D.plot(kind='scatter',x=0,y=1)
```
